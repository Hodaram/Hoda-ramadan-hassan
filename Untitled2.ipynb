```python
# %% ── CELL 1 : Install & Import Libraries ──────────────────────────────────
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time, warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, classification_report)

# تثبيت النتائج
np.random.seed(42)
tf.random.set_seed(42)

print(f"TensorFlow : {tf.__version__}")
print(f"GPU        : {len(tf.config.list_physical_devices('GPU')) > 0}")
```

```python
# %% ── Fix: تحميل CIFAR-10 من HuggingFace ───────────────────────────────────
!pip install datasets -q

from datasets import load_dataset
import numpy as np

print("Downloading CIFAR-10 from HuggingFace …")
ds = load_dataset("cifar10")

X_train_raw = np.array([np.array(img) for img in ds['train']['img']])
y_train_raw = np.array(ds['train']['label'])

X_test_raw  = np.array([np.array(img) for img in ds['test']['img']])
y_test_raw  = np.array(ds['test']['label'])

print(f"✅ Done!")
print(f"Train: {X_train_raw.shape}  |  Test: {X_test_raw.shape}")
```

```python
# %% ── CELL 3 : Visualize Sample Images ─────────────────────────────────────
fig, axes = plt.subplots(2, 5, figsize=(14, 6))
fig.suptitle('CIFAR-10 – Sample Images (Original 32×32)', fontsize=15, fontweight='bold')

for i, ax in enumerate(axes.flat):
    class_indices = np.where(y_train_raw == i)[0]

    if len(class_indices) > 0:
        idx = np.random.choice(class_indices)
        ax.imshow(X_train_raw[idx])
        ax.set_title(CLASS_NAMES[i], fontsize=11)
    else:
        ax.set_title(f"{CLASS_NAMES[i]} (No Data)")

    ax.axis('off')

plt.tight_layout()
plt.savefig('sample_images.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# %% ── CELL 4 : Preprocessing — Resize 32×32 → 224×224 + Normalize ──────────
TARGET_SIZE   = (224, 224)

# ⚠️ Colab Free  →  5000 / 1000
# Full dataset   →  change both to None
SUBSET_TRAIN  = 5000
SUBSET_TEST   = 1000

def resize_dataset(X, y, n_samples=None):
    if n_samples is not None:   # ✅ تعديل مهم
        idx = np.random.choice(len(X), n_samples, replace=False)
        X, y = X[idx], y[idx]

    X_resized = tf.image.resize(X, TARGET_SIZE).numpy().astype('float32')
    X_prep    = preprocess_input(X_resized)  # VGG16 normalization
    return X_prep, y

print("Resizing train …")
t0 = time.time()
X_train, y_train = resize_dataset(X_train_raw, y_train_raw, SUBSET_TRAIN)
print(f"  Done {time.time()-t0:.1f}s  |  shape: {X_train.shape}")

print("Resizing test …")
t0 = time.time()
X_test, y_test = resize_dataset(X_test_raw, y_test_raw, SUBSET_TEST)
print(f"  Done {time.time()-t0:.1f}s  |  shape: {X_test.shape}")

print(f"\n✅ Pixel range after preprocess_input : [{X_train.min():.1f}, {X_train.max():.1f}]")
```

```python
# %% ── CELL 5 : Train / Validation Split (80 / 20) ──────────────────────────
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train,
    test_size=0.20, stratify=y_train, random_state=42
)

y_tr_oh   = to_categorical(y_tr,   NUM_CLASSES)
y_val_oh  = to_categorical(y_val,  NUM_CLASSES)
y_test_oh = to_categorical(y_test, NUM_CLASSES)

print(f"Train : {X_tr.shape[0]}  |  Val : {X_val.shape[0]}  |  Test : {X_test.shape[0]}")
```

```python
# %% ── CELL 6 : Data Augmentation ───────────────────────────────────────────
augmentor = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1,
    fill_mode='nearest'
)

# Visualise augmented samples
fig, axes = plt.subplots(2, 5, figsize=(14, 6))
fig.suptitle('Augmented Samples (224×224)', fontsize=13, fontweight='bold')

sample_img = X_tr[:1]
aug_gen    = augmentor.flow(sample_img, batch_size=1)

# mean values used in VGG16 preprocessing (BGR)
VGG_MEAN = [103.939, 116.779, 123.68]

for ax in axes.flat:
    img = next(aug_gen)[0]

    # reverse preprocessing for display
    img_disp = np.clip(img + VGG_MEAN, 0, 255).astype(np.uint8)
    img_disp = img_disp[..., ::-1]  # BGR → RGB

    ax.imshow(img_disp)
    ax.axis('off')

plt.tight_layout()
plt.savefig('augmented_samples.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# %% ── CELL 7 : Load VGG16 Base ─────────────────────────────────────────────
vgg_base = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

print(f"\nVGG16 base parameters : {vgg_base.count_params():,}")


# %% ── CELL 8a : APPROACH 1 — VGG16 Feature Extraction ──────────────────────
feature_extractor = Model(inputs=vgg_base.input, outputs=vgg_base.output)
feature_extractor.trainable = False

def extract_features(X, model, batch_size=32):
    features = []
    for i in range(0, len(X), batch_size):
        batch = X[i:i+batch_size]
        feat  = model.predict(batch, verbose=0)
        feat  = feat.reshape(feat.shape[0], -1)   # flatten 7×7×512 = 25088
        features.append(feat)
    return np.vstack(features)

print("Extracting features — train …")
t0 = time.time()
X_tr_feat = extract_features(X_tr, feature_extractor)
print(f"  {time.time()-t0:.1f}s  |  shape: {X_tr_feat.shape}")

print("Extracting features — val …")
X_val_feat = extract_features(X_val, feature_extractor)

print("Extracting features — test …")
X_test_feat = extract_features(X_test, feature_extractor)

# ✅ تأكيد إن كل حاجة تمام
print(f"\nTrain features shape : {X_tr_feat.shape}")
print(f"Val features shape   : {X_val_feat.shape}")
print(f"Test features shape  : {X_test_feat.shape}")
```

```python
# %% ── CELL 8b : Scale + PCA ─────────────────────────────────────────────────
scaler = StandardScaler()

# Scaling
X_tr_scaled   = scaler.fit_transform(X_tr_feat)
X_val_scaled  = scaler.transform(X_val_feat)
X_test_scaled = scaler.transform(X_test_feat)

# PCA (95% variance)
pca = PCA(n_components=0.95, random_state=42)

X_tr_pca   = pca.fit_transform(X_tr_scaled)
X_val_pca  = pca.transform(X_val_scaled)
X_test_pca = pca.transform(X_test_scaled)

# معلومات مهمة
print(f"PCA components : {X_tr_pca.shape[1]}")
print(f"Explained variance retained : {pca.explained_variance_ratio_.sum()*100:.2f}%")

# تأكيد الأشكال
print(f"Train PCA shape : {X_tr_pca.shape}")
print(f"Val PCA shape   : {X_val_pca.shape}")
print(f"Test PCA shape  : {X_test_pca.shape}")
```

```python
# %% ── CELL 8c : Train SVM Classifier (FIXED) ────────────────────────────────
SVM_KERNEL = 'linear'
SVM_C      = 1.0

svm_clf = SVC(kernel=SVM_KERNEL, C=SVM_C, random_state=42)

print(f"Training SVM (kernel={SVM_KERNEL}, C={SVM_C}) …")
t0 = time.time()

try:
    svm_clf.fit(X_tr_pca, y_tr)
    svm_train_time = time.time() - t0
    print(f"  Done in {svm_train_time:.1f}s")

    # Prediction
    y_pred_svm = svm_clf.predict(X_test_pca)

    # Metrics
    acc_svm  = accuracy_score(y_test, y_pred_svm)
    prec_svm = precision_score(y_test, y_pred_svm, average='macro', zero_division=0)
    rec_svm  = recall_score(y_test, y_pred_svm, average='macro', zero_division=0)
    f1_svm   = f1_score(y_test, y_pred_svm, average='macro', zero_division=0)
    cm_svm   = confusion_matrix(y_test, y_pred_svm)

    print(f"\n── Approach 1 (VGG16 + SVM) ──────────────────")
    print(f"  Accuracy  : {acc_svm*100:.2f}%")
    print(f"  Precision : {prec_svm*100:.2f}%")
    print(f"  Recall    : {rec_svm*100:.2f}%")
    print(f"  F1-Score  : {f1_svm*100:.2f}%")
    print(f"  Time      : {svm_train_time:.1f}s")

except Exception as e:
    print("❌ SVM failed:", e)
```

```python
# %% ── CELL 8d : Confusion Matrix — Approach 1 (SAFE) ───────────────────────
if 'cm_svm' in globals() and 'y_pred_svm' in globals():

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues',
                xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)

    plt.title('Confusion Matrix – Approach 1: VGG16 + SVM',
              fontsize=13, fontweight='bold')

    plt.xlabel('Predicted')
    plt.ylabel('True')

    plt.tight_layout()
    plt.savefig('cm_approach1_svm.png', dpi=150, bbox_inches='tight')
    plt.show()

    print(classification_report(y_test, y_pred_svm, target_names=CLASS_NAMES))

else:
    print("⚠️ SVM results not found. Run SVM cell first.")
```

```python
# %% ── CELL 9 : Train + Fine-Tune + Evaluate (كل حاجة في cell واحدة) ────────
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time

import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.datasets import cifar10

from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                              f1_score, confusion_matrix, classification_report)
from sklearn.model_selection import train_test_split

# ── Settings ─────────────────────────────────────────────────────────────────
CLASS_NAMES   = ['airplane','automobile','bird','cat','deer',
                 'dog','frog','horse','ship','truck']
NUM_CLASSES   = 10
TARGET_SIZE   = (224, 224)
SUBSET_TRAIN  = 5000
SUBSET_TEST   = 1000
LEARNING_RATE = 1e-4
BATCH_SIZE    = 32
EPOCHS        = 20
DROPOUT_RATE  = 0.5

# ── Load & Resize Data ────────────────────────────────────────────────────────
(X_train_raw, y_train_raw), (X_test_raw, y_test_raw) = cifar10.load_data()
y_train_raw = y_train_raw.flatten()
y_test_raw  = y_test_raw.flatten()

def resize_dataset(X, y, n_samples=None):
    if n_samples:
        idx = np.random.choice(len(X), n_samples, replace=False)
        X, y = X[idx], y[idx]
    X_resized = tf.image.resize(X, TARGET_SIZE).numpy()
    return preprocess_input(X_resized), y

print("Resizing train …")
X_train, y_train = resize_dataset(X_train_raw, y_train_raw, SUBSET_TRAIN)
print("Resizing test  …")
X_test,  y_test  = resize_dataset(X_test_raw,  y_test_raw,  SUBSET_TEST)

X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train, test_size=0.20, stratify=y_train, random_state=42)

y_tr_oh  = to_categorical(y_tr,  NUM_CLASSES)
y_val_oh = to_categorical(y_val, NUM_CLASSES)

print(f"Train:{X_tr.shape[0]}  Val:{X_val.shape[0]}  Test:{X_test.shape[0]}")

# ── Augmentation ──────────────────────────────────────────────────────────────
augmentor = ImageDataGenerator(
    rotation_range=15, width_shift_range=0.1, height_shift_range=0.1,
    horizontal_flip=True, zoom_range=0.1, fill_mode='nearest')

# ── Build Model ───────────────────────────────────────────────────────────────
base = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
for layer in base.layers:
    layer.trainable = False

e2e_model = Sequential([
    base,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(DROPOUT_RATE),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(NUM_CLASSES, activation='softmax')
], name='VGG16_EndToEnd')

e2e_model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
# CELL 9b ── Train Frozen
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)]

train_gen = augmentor.flow(X_tr, y_tr_oh, batch_size=BATCH_SIZE, shuffle=True)

print(f"\n── Phase 1: Training (Frozen) ──────────────────────────────────────")
t0 = time.time()
history = e2e_model.fit(
    train_gen,
    steps_per_epoch=len(X_tr) // BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(X_val, y_val_oh),
    callbacks=callbacks,
    verbose=1)
e2e_train_time = time.time() - t0
print(f"Done in {e2e_train_time:.1f}s ({e2e_train_time/60:.1f} min)")
```

```python
# CELL 9b-2 ── Fine-Tuning (فك آخر 4 layers)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
print(f"\n── Phase 2: Fine-Tuning (unfreeze last 4 layers) ───────────────────")
for layer in e2e_model.layers[0].layers[-4:]:
    layer.trainable = True

e2e_model.compile(
    optimizer=Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy'])

callbacks_ft = [
    EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True, verbose=1)]

t0 = time.time()
history_ft = e2e_model.fit(
    X_tr, y_tr_oh,
    batch_size=BATCH_SIZE,
    epochs=10,
    validation_data=(X_val, y_val_oh),
    callbacks=callbacks_ft,
    verbose=1)
ft_train_time = time.time() - t0
print(f"Fine-tuning done in {ft_train_time:.1f}s ({ft_train_time/60:.1f} min)")
```

```python
# CELL 9c ── Learning Curves
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Frozen training curves
ep1 = range(1, len(history.history['accuracy']) + 1)
axes[0,0].plot(ep1, history.history['accuracy'],     'b-o', label='Train')
axes[0,0].plot(ep1, history.history['val_accuracy'], 'r-s', label='Val')
axes[0,0].set_title('Accuracy – Frozen Training', fontweight='bold')
axes[0,0].legend(); axes[0,0].grid(True)

axes[0,1].plot(ep1, history.history['loss'],     'b-o', label='Train')
axes[0,1].plot(ep1, history.history['val_loss'], 'r-s', label='Val')
axes[0,1].set_title('Loss – Frozen Training', fontweight='bold')
axes[0,1].legend(); axes[0,1].grid(True)

# Fine-tuning curves
ep2 = range(1, len(history_ft.history['accuracy']) + 1)
axes[1,0].plot(ep2, history_ft.history['accuracy'],     'b-o', label='Train')
axes[1,0].plot(ep2, history_ft.history['val_accuracy'], 'r-s', label='Val')
axes[1,0].set_title('Accuracy – Fine-Tuning', fontweight='bold')
axes[1,0].legend(); axes[1,0].grid(True)

axes[1,1].plot(ep2, history_ft.history['loss'],     'b-o', label='Train')
axes[1,1].plot(ep2, history_ft.history['val_loss'], 'r-s', label='Val')
axes[1,1].set_title('Loss – Fine-Tuning', fontweight='bold')
axes[1,1].legend(); axes[1,1].grid(True)

plt.tight_layout()
plt.savefig('learning_curves_e2e.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ saved → learning_curves_e2e.png")
```

```python
# CELL 9d ── Evaluate
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
y_pred_e2e = np.argmax(e2e_model.predict(X_test, verbose=0), axis=1)

acc_e2e  = accuracy_score(y_test, y_pred_e2e)
prec_e2e = precision_score(y_test, y_pred_e2e, average='macro')
rec_e2e  = recall_score(y_test, y_pred_e2e, average='macro')
f1_e2e   = f1_score(y_test, y_pred_e2e, average='macro')
cm_e2e   = confusion_matrix(y_test, y_pred_e2e)

print(f"\n── Approach 2 Results (After Fine-Tuning) ──────────────────────────")
print(f"  Accuracy  : {acc_e2e*100:.2f}%")
print(f"  Precision : {prec_e2e*100:.2f}%")
print(f"  Recall    : {rec_e2e*100:.2f}%")
print(f"  F1-Score  : {f1_e2e*100:.2f}%")
print(f"  Time      : {e2e_train_time + ft_train_time:.1f}s")

plt.figure(figsize=(10, 8))
sns.heatmap(cm_e2e, annot=True, fmt='d', cmap='Greens',
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.title('Confusion Matrix – Approach 2: VGG16 End-to-End + Fine-Tuning',
          fontsize=13, fontweight='bold')
plt.xlabel('Predicted'); plt.ylabel('True')
plt.tight_layout()
plt.savefig('cm_approach2_e2e.png', dpi=150, bbox_inches='tight')
plt.show()
print(classification_report(y_test, y_pred_e2e, target_names=CLASS_NAMES))
```

```python
# %% ── CELL 10a : Comparative Bar Chart ──────────────────────────────────────
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
import warnings
warnings.filterwarnings('ignore')

# ── القيم الثابتة — عدّلها لو عندك نتايج مختلفة ──
acc_svm       = globals().get('acc_svm',  0.0)
prec_svm      = globals().get('prec_svm', 0.0)
rec_svm       = globals().get('rec_svm',  0.0)
f1_svm        = globals().get('f1_svm',   0.0)
acc_e2e       = globals().get('acc_e2e',  0.0)
prec_e2e      = globals().get('prec_e2e', 0.0)
rec_e2e       = globals().get('rec_e2e',  0.0)
f1_e2e        = globals().get('f1_e2e',   0.0)
svm_train_time = globals().get('svm_train_time', 0.0)
e2e_train_time = globals().get('e2e_train_time', 0.0)

metrics  = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
vals_svm = [acc_svm,  prec_svm,  rec_svm,  f1_svm]
vals_e2e = [acc_e2e,  prec_e2e,  rec_e2e,  f1_e2e]

x, width = np.arange(len(metrics)), 0.35
fig, ax = plt.subplots(figsize=(10, 6))

b1 = ax.bar(x - width/2, [v*100 for v in vals_svm], width,
            label='Approach 1: VGG16 + SVM',
            color='steelblue', edgecolor='white')
b2 = ax.bar(x + width/2, [v*100 for v in vals_e2e], width,
            label='Approach 2: VGG16 End-to-End',
            color='darkorange', edgecolor='white')

ax.set_xlabel('Metric', fontsize=12)
ax.set_ylabel('Score (%)', fontsize=12)
ax.set_title('Comparative Analysis – Approach 1 vs Approach 2',
             fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(metrics, fontsize=11)
ax.set_ylim(0, 110)
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.4)

for bar in list(b1) + list(b2):
    ax.annotate(f'{bar.get_height():.1f}%',
                xy=(bar.get_x() + bar.get_width()/2, bar.get_height()),
                xytext=(0, 4), textcoords='offset points',
                ha='center', fontsize=9)

plt.tight_layout()
plt.savefig('comparison_bar_chart.png', dpi=150, bbox_inches='tight')
plt.show()
```
